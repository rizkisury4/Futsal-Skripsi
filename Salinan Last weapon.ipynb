```python
# Import library
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
from mplsoccer import Pitch
from collections import defaultdict
import os

# 1. Load Model YOLO
model = YOLO('yolov8n.pt')  # Load YOLOv8 detection model

# 2. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 3. Folder Output
output_folder = "/content/drive/MyDrive/monyet"
os.makedirs(output_folder, exist_ok=True)

# 4. Input Video
input_video_path = '/content/drive/MyDrive/monyet/datauu.mp4'  # Replace with your video path
output_video_path = os.path.join(output_folder, 'trajectory_output.mp4')

# 5. Initialize variables for trajectory tracking
trajectories = defaultdict(list)

# 6. Video Capture
cap = cv2.VideoCapture(input_video_path)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

# 7. Process video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform YOLO prediction on the current frame
    results = model(frame)
    detections = results[0].boxes

    # Draw detections and update trajectories
    for box in detections:
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())  # Bounding box coordinates
        obj_id = int(box.cls)  # Class ID (e.g., players, ball)
        conf = box.conf[0].item()  # Confidence score

        # Update trajectory
        center_x, center_y = (x1 + x2) // 2, (y1 + y2) // 2
        trajectories[obj_id].append((center_x, center_y))

        # Draw bounding box and label
        color = (0, 255, 0)  # Green for players
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(frame, f'ID: {obj_id} {conf:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Draw trajectories on the frame
    for obj_id, trajectory in trajectories.items():
        for i in range(1, len(trajectory)):
            cv2.line(frame, trajectory[i - 1], trajectory[i], (255, 0, 0), 2)  # Blue line for trajectory

    # Save processed frame to output video
    out.write(frame)

# Release resources
cap.release()
out.release()




print(f"Output video saved at: {output_video_path}")
```

```python
# Import library
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from collections import defaultdict
import os

# 3. Folder Output
output_folder = "/content/drive/MyDrive/monyet"
os.makedirs(output_folder, exist_ok=True)

# 4. Input Video
input_video_path = '/content/drive/MyDrive/monyet/datauu.mp4'  # Replace with your video path
output_video_path = os.path.join(output_folder, 'trajectory_output.mp4')

# 5. Path to Futsal Court Background
background_path = '/content/drive/MyDrive/dataset/futsal-court-or-field-sport-background-line-art-style-vector.jpg'
background = cv2.imread(background_path)

# Resize background to match video dimensions
cap = cv2.VideoCapture(input_video_path)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
background = cv2.resize(background, (frame_width, frame_height))

# 6. Initialize variables for trajectory tracking
trajectories = defaultdict(list)

# Video writer
out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

# DataFrame for trajectory data
trajectory_data = []

# 7. Process video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform YOLO prediction on the current frame
    results = model(frame)
    detections = results[0].boxes

    # Create a copy of the resized background
    output_frame = background.copy()

    # Draw detections and update trajectories
    for box in detections:
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())  # Bounding box coordinates
        obj_id = int(box.cls)  # Class ID
        conf = box.conf[0].item()  # Confidence score

        # Update trajectory
        center_x, center_y = (x1 + x2) // 2, (y1 + y2) // 2
        trajectories[obj_id].append((center_x, center_y))
        trajectory_data.append({'frame': cap.get(cv2.CAP_PROP_POS_FRAMES),
                                'object_id': obj_id,
                                'x': center_x,
                                'y': center_y})

    # Draw trajectories on the background
    for obj_id, trajectory in trajectories.items():
        for i in range(1, len(trajectory)):
            cv2.line(output_frame, trajectory[i - 1], trajectory[i], (255, 0, 0), 2)  # Blue line for trajectory

    # Save processed frame to output video
    out.write(output_frame)

# Release resources
cap.release()
out.release()

# 8. Save trajectory data to Excel
df = pd.DataFrame(trajectory_data)
excel_path = "/content/drive/MyDrive/dataset/trajectories.xlsx"
df.to_excel(excel_path, index=False)

print(f"Output video saved at: {output_video_path}")
print(f"Trajectory data saved at: {excel_path}")
```
